```python
from math import exp
from random import random
#activation function
def sigmoid(activate):
    return 1.0 / (1.0 + exp(-activate))

def sigmoid_derivative(activate):
    temp=sigmoid(activate)
    return temp*(1-temp)

def relu(activate):
    return max(0,activate)

def relu_derivative(activate):
    if activate>=0:
        return 1
    else:
        return 0
    
def activate(weights, inputs):
    #마지막 weight는 bias로 설정 했기 때문에 맨처음에 더해준다.
    activation = weights[-1]
    for i in range(len(weights)-1):
        activation += weights[i] * inputs[i]
    return activation
 
    
def make_model(hl_shape, input_shape, output_shape):
    model=list()
    #input_shape+1인 거는 bias를 추가했기 때문이다.
    #hidden_layer = [{'weights':[random() for _ in range(input_shape + 1)]} for _ in range(hl_shape)]
    
    #hidden_layer = [{'weights':[random() for _ in range(input_shape + 1)]} for _ in range(hl_shape)]
    #model.append(hidden_layer)
    
    #output_layer도 bias
    #output_layer=[{'weights':[random() for _ in range(hl_shape + 1)]} for _ in range(output_shape)]
    #model.append(output_layer)
    
    hidden_layer=[]
    output_layer=[]
    for _ in range(hl_shape):
        hidden_layer.append({'weights':list(np.random.normal(0,1,input_shape))})
    model.append(hidden_layer)
    
    for _ in range(output_shape):
        output_layer.append({'weights':list(np.random.normal(0,1,hl_shape))})
    model.append(output_layer)
    
    return model

def forward(network, inputs):
    temp=inputs
    for num, layer in enumerate(network):
        new_inputs=[]
        for neroun in layer:
            activater=activate(neroun['weights'],temp)
            #neroun['output']=sigmoid(activater)
            
            if num==len(network)-1:
                neroun['output']=sigmoid(activater)
                #print(neroun)
            else:
                neroun['output']=relu(activater)
           
            new_inputs.append(neroun['output'])
        temp=new_inputs
    return temp
def backward(model,target):
    cost=0
    for i in range(len(model)-1,-1,-1):
        layer=model[i]
        if i==len(model)-1:
            for n in range(len(layer)):
                neroun=layer[n]
                temp=neroun['output']-target[n]
                #errors.append(temp)
                neroun['delta']= temp*sigmoid_derivative(neroun['output'])
                cost+=(temp**2)
        else:
            errors=list()
            for n in range(len(layer)):
                error = 0
                for neuron in model[i + 1]:
                    #다음 layer에 있는 neuron들의 delta값을 가지고 와 weight와 곱하고 다 더해준다.
                    error += (neuron['weights'][n] * neuron['delta'])
                errors.append(error)
            for n in range(len(layer)):
                neroun=layer[n]
                neroun['delta']= errors[n]*relu_derivative(neroun['output'])
    return cost

def update_weights(model, one_train, one_label, lr):
    for i in range(len(model)):
        inputs=one_train
        if i!=0:
            inputs=[neroun['output'] for neroun in model[i-1]]
        for neuron in model[i]:
            for n in range(len(inputs)):
                neuron['weights'][n] -= lr * neuron['delta'] * inputs[n]
            neuron['weights'][-1] -= lr * neuron['delta']
            
def train_model(model, x_train, y_train, lr, n_epoch, n_outputs):
    for epoch in range(n_epoch):
        sum_error = 0
        for row in range(len(x_train)):
            outputs = forward(model, x_train[row])
            expected = [0 for _ in range(n_outputs)]
            expected[y_train[row]] = 1
            sum_error=backward(model, expected)
            update_weights(model, x_train[row],y_train[row], lr)
        print('-epoch=%d, lrate=%.3f, error=%.3f' % (epoch+1, lr, sum_error))
        
def predict(model, x_test, y_test):
    cnt=0
    for n in range(len(x_test)):
        output=forward(model,x_test[n])
        model_predict=max(output)
        if (output.index(model_predict))== y_test[n]:
             cnt+=1
    return cnt/len(x_test)*100
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data, label, test_size=0.2, shuffle=True, random_state=42)
model=[]
model=make_model(3,12,2)
train_model(model,x_train,y_train,0.2, 10, 2)
predict_rate=predict(model, x_test,y_test)
print(predict_rate)
print(model)
```
